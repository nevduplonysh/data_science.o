### Convergence of relative frequency to probability

Let's see how this works using a coin toss as an example.

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=42)
plt.style.use('dark_background')
```

```python
# flip simulation
n_flips = 10_000

flips = rng.binomial(n=1, p=0.5, size=n_flips)
print(flips[:20])

cumulative_heads = np.cumsum(flips) # heads
print(cumulative_heads)

n_range = np.arange(1, n_flips + 1) # x
print(n_range)

running_frequency = cumulative_heads / n_range # y
print(running_frequency)
```

```python
def show(n_range, running_frequency, p):
    plt.figure(figsize=(10, 5))
    plt.plot(n_range, running_frequency, label="N(A) / A")
    plt.axhline(y=p, color="red", linestyle="--", label="P(A) = 0.5")
    plt.xscale("log")
    plt.xlabel("Number of tosses (N)")
    plt.ylabel("Frequency of heads")
    plt.title("Convergence of frequency to probability")
    plt.legend()
    plt.grid(True, alpha=0.15)
    plt.show()
```

```python
show(n_range, running_frequency, p=0.5)
```

We can see how our coin-tossing frequency converges to p=0.5 as the number of tosses grows.

We can observe the same pattern if we change p=0.5 to p=0.3, as shown below:

```python
flips = rng.binomial(n=1, p=0.3, size=n_flips)
print(flips[:20])

cumulative_heads = np.cumsum(flips) # heads
print(cumulative_heads)

n_range = np.arange(1, n_flips + 1) # x
print(n_range)

running_frequency = cumulative_heads / n_range # y
print(running_frequency)

show(n_range, running_frequency, p=0.3)
```
